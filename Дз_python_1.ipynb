### Задание 1 ★☆☆  
Вычисление чисел Фибоначчи с использованием рекурсии

1. Помните рекурсивную функцию fibonacci_recursive(n) с семинара, которая вычисляет n-е число Фибоначчи?
2. Используя эту функцию, напишите новую функцию, которая выведит первые n чисел Фибоначчи в список. Пусть n=20, а первое и второе числа Фибоначчи равны 0 и 1.

```python
def fibonacci_recursive(n):
    if n <= 1:
      return n

    return fibonacci_recursive(n-1) + fibonacci_recursive(n-2)
```

```python
def fibonacci_recursive_list(n):
  return [fibonacci_recursive(i) for i in range(n)]

n = 20
print(fibonacci_recursive_list(n))
```

### Задание 2 ★☆☆  
Вычисление чисел Фибоначчи с помощью цикла.
1. Вспомним функцию fibonacci_iterative(n) с семинара, которая вычисляет n-е число Фибоначчи с использованием цикла.  
2. Напишите новую функцию, которая выводит первые n чисел Фибоначчи в виде словаря, где ключом будет порядковый номер числа Фибоначчи, а значением -  само число Фиббоначи. Пусть n=20, а первое и второе числа Фибоначчи равны 0 и 1.
3. Далее переверните словарь так, чтобы там были последние 15 из 20 чисел Фибоначчи, где ключ - это значение числа Фибоначчи, а значение - порядковый номер этого числа.

```python
def fibonacci_iterative(n):
    if n <= 1:
      return n
    i, j = 0, 1

    for _ in range(2, n):
      i, j = j, i+j

    return j

def fibonacci_iterative_dict(n):
  return  {i: fibonacci_iterative(i) for i in range(n+1)}


n = 20
print(f'Первые 20 чисел Фибоначчи в виде словаря {fibonacci_iterative_dict(n)}')


new_dict = {v:k for k, v in fibonacci_iterative_dict(n).items() if 15 <= k <= 20}
print(f'Последние 15 из 20 чисел Фибоначчи в виде перевернутого словаря {new_dict}')
```

### Задание 3 ★☆☆  
Сумма четных чисел в последовательности с условиями

Напишите функцию conditional_even_sum(sequence, threshold), которая принимает список целых чисел и возвращает сумму всех четных чисел в этом списке, которые больше заданного порогового значения threshold.  

`sequence = [1, 2, 3, 4, 5, 6, 8, 9, 12, 15, 20]`

`threshold = 5`

```python
def conditional_even_sum(sequence, threshold):
    return sum(v for v in sequence if v % 2 == 0 and v > threshold)

sequence = [1, 2, 3, 4, 5, 6, 8, 9, 12, 15, 20]

threshold = 5

print(conditional_even_sum(sequence, threshold))
```

### Задание 4 ★★☆  
Сумма произведений уникальных пар чисел

Напишите функцию sum_of_unique_products(sequence), которая принимает список целых чисел и возвращает сумму произведений всех уникальных пар чисел из списка. Пара (a, b) и (b, a) считаются одинаковыми и должны учитываться только один раз.  
`sequence = [1, 2, 3]`


```python
def sum_of_unique_products(sequence):
    sum = 0
    for i in range(len(sequence)):
      for j in range(i+1, len(sequence)):
        val = sequence[i] * sequence[j]
        sum+=val

    return sum

sequence = [1, 2, 3]
print(sum_of_unique_products(sequence))
```

Напишите функцию is_palindrome(s), которая принимает строку s и возвращает True, если строка является палиндромом, и False в противном случае. Строка считается палиндромом, если она читается одинаково слева направо и справа налево, игнорируя пробелы, знаки препинания и регистр символов.  
`s = "A man, a plan, a canal, Panama!"`


```python
def is_palindrome(s):
   return ''.join([c.lower() for c in s if c.isalpha()])[::1]==''.join([c.lower() for c in s if c.isalpha()])[::-1]


s1= "A man, a plan, a canal, Panama!"
print(is_palindrome(s1))

s2= "A man, a plan, a canal, Panama, Panama!"
print(is_palindrome(s2))
```

### Задание 7 ★☆☆ (типы данных)
Счетчик целых чисел

Напишите функцию dict_counter, которая принимает на вход список целых чисел (sequence) и возвращает словарь, ключи которого представляют собой уникальные числа последовательности, а значения - их количество.

```python
def dict_counter(sequence):
  dict_count = {}
  for i in sequence:
    if i in dict_count:
      dict_count[i] += 1
    else:
       dict_count[i] = 1
  return dict_count

print(dict_counter([1,2,2,3,3,3,4,4]))
```

### Задание 8 ★☆☆ (на map и filter)
Фильтрация списков

Напишите функцию filter_numbers, которая принимает на вход список целых чисел (sequence), затем, не используя циклы, умножает их на -1 и возвращает только положительные получившиеся значения.
Разобраться с тонкостимя работы map и filter нужно самостоятельно.

```python
def filter_numbers(sequence):
  list_numbers = list(map(lambda x: x * -1, sequence))
  result = list(filter(lambda x: x > 0, list_numbers))
  return result

sequence = [1, -2, 3, -4, 5, -6]

print(filter_numbers(sequence))
```

### Задание 9 ★★☆ (списки, сортировка)
Сортировка отрезков по длине

Напишите функцию sorted_segments, которая принимает на вход список отрезков (sequence) и сортирует их по увеличению длины. Отрезок представляет собой кортеж из пары чисел (a, b), a <= b, а его длина - разница b и a.

Подсказка: у функций sorted и sort есть необязательный параметр key, который можно задать для условия сортировки.

```python
def lenght(segment):
  return segment[1]-segment[0]


def sorted_segments(sequence):
  return sorted(sequence, key=lenght)

sequence = [(1,2), (3,8), (2, 10), (1, 3), (7, 12)]

print(sorted_segments(sequence))
```
